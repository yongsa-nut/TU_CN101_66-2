![onebitmatmul.png](https://drive.google.com/uc?export=view&id=1g_YCl3R_uif0sTAWbms1x2mhVwENPNqv)

There are two steps in this program.

1.   Turn FP16 matrixes into 1.58bit matrix.

     To constrain the weights to -1, 0, or +1, we adopt an absmean quantization function. It first scales the weight matrix by its average absolute value, and then round each value to the nearest integer among {-1, 0, +1}.

    $M_{i,j} = max(-1, min(1, round(\frac{M_{i,j}}{\gamma + ϵ})))$

    $\gamma = \frac{1}{nm}\sum_{i,j}|M_{i,j}|$

    ,where $M$ is a $n \times m$ matrix, $\gamma$ is the scaling factor, and $ϵ$ is a small number to add for stability when $\gamma$ is too small.
